## Text detection from images using EasyOCR : Hands on guide

### What is OCR?
OCR is formely known as Optical Character Recognition which is revolutionary for a digital world nowdays. OCR is actually a complete process under which the images/documents which are present in a digital world are processed and from them text is being processed out as a normal text.
### Purpose of OCR
OCR is a technology that enables you to convert different types of documents, such as scanned paper documents, PDF files or images captured by a digital camera into editable and searchable data.

### What is EasyOCR?
EasyOCR is actually a python package which hold pytorch as a backend handler. EasyOCR like any other OCR(tesseract of Google or any other) detects the text from images but in my reference while using it I found that it is the most straightforward way to detect text from images also when high end deep learning library(pytorch) is supporting it in the backend which makes it accuracy more credible. EasyOCR supports 42+ languages for detection purpose. EasyOCR is created by the company named Jaided AI company.

### 5.1. Draw results for single line text - Example 1

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

In the above code snippet,
1. We are trying to get the coordinates to draw the bounding box and text over our image on which we have perform our detection.
2. In top_left varible we are getting the coordinate of top left corner in the form of tuple accessing from results. Similarly we can see that in bottom right coordinate.

```python
img
```

```python
img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,bottom_right, font, 0.5,(0,255,0),2,cv2.LINE_AA)
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```

### 5.3. Draw results for multiple lines

```python
img = cv2.imread('/content/IJCIS-12-2-1575-g002.png')
spacer = 100
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15

plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```

In the above code snippet, we just need to focus on few points:
1. Instead for detecting one line text, here we are looping through all the detection as we want to plot multiple lines text
2. While giving the coordinates on cv2.putText we are using an extra variable which is "spacer" this spacer later in the code is being incremented to +15 which is helping to restrict the text to collide over each other.
3. This spacer variable will help the text to remain sorted and equally spaced.
